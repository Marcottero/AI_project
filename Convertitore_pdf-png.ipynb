```python
from IPython import get_ipython
from IPython.display import display
# %%
!pip install PyMuPDF pillow
# %%
from google.colab import drive

drive.mount('/content/gdrive')
# %%
import fitz  # PyMuPDF
import os

def render_pages_to_images(pdf_path, output_folder, dpi=300):
    """
    Renderizza ogni pagina di un PDF in un'immagine PNG.

    Args:
        pdf_path (str): Il percorso del file PDF.
        output_folder (str): La cartella dove salvare le immagini PNG.
        dpi (int): La risoluzione (dots per inch) per il rendering.
                   Un valore più alto produce un'immagine più grande e dettagliata.
    Returns:
        int: Il numero di immagini create.
    """
    # Apri il PDF
    try:
        pdf_document = fitz.open(pdf_path)
        print(f"PDF aperto con successo: {pdf_path}")
    except Exception as e:
        print(f"Errore nell'apertura del PDF {pdf_path}: {e}")
        return 0

    # Crea la cartella output se non esiste
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
        print(f"Creata la cartella di output: {output_folder}")

    image_count = 0
    # Itera attraverso ogni pagina
    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        print(f"Rendering pagina {page_num + 1} con DPI={dpi}")

        try:
            # Renderizza la pagina come un'immagine (pixmap) con la risoluzione specificata
            pix = page.get_pixmap(dpi=dpi)

            # Genera un nome file per la pagina renderizzata
            # Assicurati che il nome file sia compatibile (senza caratteri non validi)
            base_filename = os.path.basename(pdf_path).split('.')[0]
            image_filename = f"{base_filename}_page_{page_num+1}.jpg"
            image_path = os.path.join(output_folder, image_filename)

            # Salva il pixmap come file immagine
            pix.save(image_path)
            image_count += 1
            print(f"Pagina renderizzata e salvata come immagine: {image_path}")

        except Exception as e:
            print(f"Errore nel rendering o salvataggio della pagina {page_num + 1}: {e}")


    pdf_document.close()
    return image_count

# Cartella contenente i PDF
pdf_folder = "/content/gdrive/My Drive/Colab Notebooks/Dataset_V2"
# Cartella di output per le immagini renderizzate delle pagine
output_folder = "/content/gdrive/My Drive/Colab Notebooks/Dataset_V2.1_Renderedjpg"

print(f"Controllo dei file nella cartella: {pdf_folder}")
# Elabora tutti i PDF nella cartella
total_images_rendered = 0
try:
    pdf_files_in_folder = [f for f in os.listdir(pdf_folder) if f.endswith('.pdf')]
    print(f"Trovati {len(pdf_files_in_folder)} file PDF nella cartella.")

    for pdf_file in pdf_files_in_folder:
        pdf_path = os.path.join(pdf_folder, pdf_file)
        print(f"Elaborazione del file: {pdf_file}")
        # Puoi regolare il DPI qui per cambiare la qualità dell'immagine renderizzata
        images_rendered = render_pages_to_images(pdf_path, output_folder, dpi=300)
        total_images_rendered += images_rendered
        print(f"Renderizzate {images_rendered} immagini di pagina dal file {pdf_file}")

    print(f"Totale immagini di pagina renderizzate: {total_images_rendered}")
except FileNotFoundError:
    print(f"Errore: La cartella dei PDF non è stata trovata: {pdf_folder}")
except Exception as e:
    print(f"Si è verificato un errore generale: {e}")
```
